```python
# import library
from pytube import YouTube
```

## Quick Code ...

### ... for Videos:

```python
video_url = "https://www.youtube.com/watch?time_continue=7&v=AxWS5cz2Lt0"
SAVE_PATH = "/Users/dirkkalmbach/Downloads" # <- folder
yt = YouTube(video_url)
yt.streams.filter(progressive=True)
stream = yt.streams.get_by_itag(22)#seems: 22=always 720p
stream.download(SAVE_PATH)
```

### ... for Audio:

```python
t=yt.streams.filter(only_audio=True)
t[0].download(SAVE_PATH)
```

### Thumbnail ðŸŽ‘

```python
# get thumbnail url
yt.thumbnail_url
```

```python
# Download thumbnail image
import requests
import os
img_file = os.path.join(SAVE_PATH, "thumbnail.jpg")
url = yt.thumbnail_url
response = requests.get(url)
with open(img_file, "wb") as f:
    f.write(response.content)
```

```python
# show image
from PIL import Image
Image.open(img_file)
```
